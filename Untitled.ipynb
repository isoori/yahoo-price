```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score

from keras.models import Sequential
from keras.layers import Dense,LSTM,SimpleRNN,Dropout
```

```python
df = pd.read_csv("yahoo_stock.csv")
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.set_index('Date',inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
(df['Close'] == df['Adj Close']).all()
```

```python
df.drop('Adj Close',axis=1 , inplace=True) #As Close Column and Adj Close are equal , we drop one of them 
```

```python
df.head()
```

```python
sns.barplot(x=df.index.month, y=df.Volume)
plt.title("Volume per Month")
plt.show()
```

```python
features = df.drop("Close",axis=1)
target = df[['Close']]
```

```python
features.head()
```

```python
target.head()
```

```python
train = features[:len(features)-200:]
train
```

```python
test = features[len(train):]
test
```

```python
scaller_dt = MinMaxScaler(feature_range=(0,1))
train_scaller = scaller_dt.fit_transform(train)
train_scaller
```

```python
scaller_tg = MinMaxScaler(feature_range=(0,1))
target_sc = scaller_tg.fit_transform(target)
```

```python
target_sc
```

```python
X_train , y_train = [] , []
timesteps = 50

for i in range(timesteps,len(train_scaller)):
    X_train.append(train_scaller[i-timesteps:i,:])
    y_train.append(target_sc[i,0])
```

```python
X_train = np.array(X_train)
y_train = np.array(y_train)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
model_1 = Sequential()
model_1.add(LSTM(16, activation='tanh', input_shape = (X_train.shape[1],4),return_sequences=False))
model_1.add(Dense(1))
model_1.compile(optimizer='adam',loss='mse')
```

```python
model_1.fit(X_train,y_train,batch_size=16,epochs=50)
```

```python
Test_set = features[len(features) - len(test) - timesteps :]
```

```python
t_scalled = scaller_dt.transform(Test_set)
t_scalled
```

```python
X_test = []
for i in range(timesteps,len(t_scalled)):
    X_test.append(t_scalled[i-timesteps:i,:])
```

```python
X_test = np.array(X_test)
X_test.shape
```

```python
predict_model_1 = model_1.predict(X_test)
```

```python
normal_price_1 = scaller_tg.inverse_transform(predict_model_1)
```

```python
y_test = target[-len(X_test):]
```

```python
model_1 = Sequential()
model_1.add(LSTM(10,activation='tanh',input_shape=(X_train.shape[1],4),return_sequences=False ))
model_1.add(Dense(1))
model_1.compile(optimizer='adam',loss='mse')
```

```python
model_1.fit(X_train,y_train,batch_size=16,epochs=50)
```

```python
predict_model_1 = model_1.predict(X_test)
normal_price_1 = scaller_tg.inverse_transform(predict_model_1)
```

```python
plt.figure(figsize=(20,10),dpi=100)
plt.xlabel("Days")
plt.ylabel("Value")
plt.title("Actual Stock Price vs Predicted Price for Yahoo ")
plt.plot(y_test['Close'].values,label='Actual Value ')
plt.plot(normal_price_1,label='Predicted Value For Model 1 with LSTM')
plt.legend()
plt.grid()
plt.show()
```

```python
s1 = r2_score(y_test['Close'].values,normal_price_1)
```

# model 2

```python
model_2 = Sequential()
model_2.add(LSTM(16, activation='tanh', input_shape = (X_train.shape[1],4),return_sequences=False))
model_2.add(Dense(1))
model_2.compile(optimizer='adam',loss='mse')
```

```python
model_2.fit(X_train,y_train,batch_size=32,epochs=100)
```

```python
predict_model_2 = model_2.predict(X_test)
normal_price_2 = scaller_tg.inverse_transform(predict_model_2)
```

```python
plt.figure(figsize=(20,10),dpi=100)
plt.xlabel("Days")
plt.ylabel("Value")
plt.title("Actual Stock Price vs Predicted Price for Yahoo ")
plt.plot(y_test['Close'].values,label='Actual Value ')
plt.plot(normal_price_2,label='Predicted Value For Model 2 with LSTM')
plt.legend()
plt.grid()
plt.show()
```

```python
s2= r2_score(y_test['Close'].values, normal_price_2)
```

# Model 3

```python
model_3 = Sequential()
model_3.add(LSTM(50, activation='tanh', input_shape = (X_train.shape[1],4),return_sequences=True))
model_3.add(LSTM(40, activation='tanh', input_shape = (X_train.shape[1],4),return_sequences=True))
model_3.add(LSTM(30, activation='tanh', input_shape = (X_train.shape[1],4),return_sequences=False))
model_3.add(Dense(20))
model_3.add(Dense(10))
model_3.add(Dense(1))
model_3.compile(optimizer='adam',loss='mse')
```

```python
model_3.fit(X_train,y_train,batch_size=32,epochs=100)
```

```python
predict_model_3 = model_3.predict(X_test)
normal_price_3 = scaller_tg.inverse_transform(predict_model_3)
```

```python
plt.figure(figsize=(20,10),dpi=100)
plt.xlabel("Days")
plt.ylabel("Value")
plt.title("Actual Stock Price vs Predicted Price for Yahoo ")
plt.plot(y_test['Close'].values,label='Actual Value ')
plt.plot(normal_price_3,label='Predicted Value For Model 3 with LSTM')
plt.legend()
plt.grid()
plt.show()
```

```python
s3 = r2_score(y_test['Close'].values, normal_price_3)
```

# Model 4

```python
model_4 = Sequential()
model_4.add(LSTM(50, activation='tanh', input_shape = (X_train.shape[1],4),return_sequences=True))
model_4.add(LSTM(20, activation='tanh', input_shape = (X_train.shape[1],4),return_sequences=False))
model_4.add(Dense(1))
model_4.compile(optimizer='adam',loss='mse')
```

```python
model_4.fit(X_train,y_train,batch_size=32,epochs=100)
```

```python
predict_model_4 = model_4.predict(X_test)
normal_price_4 = scaller_tg.inverse_transform(predict_model_4)
```

```python
plt.figure(figsize=(20,10),dpi=100)
plt.xlabel("Days")
plt.ylabel("Value")
plt.title("Actual Stock Price vs Predicted Price for Yahoo ")
plt.plot(y_test['Close'].values,label='Actual Value ')
plt.plot(normal_price_4,label='Predicted Value For Model 4 with LSTM')
plt.legend()
plt.grid()
plt.show()
```

```python
s4 = r2_score(y_test['Close'].values, normal_price_4)
```
